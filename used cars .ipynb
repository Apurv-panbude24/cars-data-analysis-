```python
plt.figure(figsize=(10,6))
sns.histplot(data["Kilometers_Driven"], kde=True, color='skyblue', bins=30, edgecolor='black')
plt.xlabel("Kilometers Driven", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.title("Distribution of Kilometers Driven", fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
def log_transform(data, col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + "_log"] = np.log(data[colname] + 1)
        else:
            data[colname + "_log"] = np.log(data[colname])
    data.info()
```

```python
log_transform(data, ['Kilometers_Driven','Price'])
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data["Kilometers_Driven_log"], kde=True, color='skyblue', bins=30, edgecolor='black')
plt.xlabel("Kilometers Driven_log", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.title("Distribution of Kilometers Driven", fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.pairplot(data = data.drop(['Kilometers_Driven','Price','Name','Location'], axis = 1))
plt.show()
```

```python
fig, ax = plt.subplots(4,2, figsize=(15,20))
data.groupby('Location')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax= ax[0][0])
ax[0][0].set_title('Location vs Price')
data.groupby('Transmission')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax= ax[0][1])
ax[0][1].set_title('Transmission vs Price')
data.groupby('Fuel_Type')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax= ax[1][0])
ax[1][0].set_title('Fuel_Type vs Price')
data.groupby('Owner_Type')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax= ax[1][1])
ax[1][1].set_title('Owner_Type vs Price')
data.groupby('Brand')['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax= ax[2][0])
ax[2][0].set_title('Brand vs Price')
data.groupby('Model')['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax= ax[2][1])
ax[2][1].set_title('Model vs Price')
data.groupby('Seats')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax= ax[3][0])
ax[3][0].set_title('Seats vs Price')
data.groupby('Age')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax= ax[3][1])
ax[3][1].set_title('Age vs Price')
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=0.5)
sns.despine()
## with price set as with log 
```
